```python
def total_range(melody_to_analyze_df):

    min_note = note_conversion_df.Order[melody_to_analyze_df.Bir.iloc[0]]
    max_note = note_conversion_df.Order[melody_to_analyze_df.Bir.iloc[0]]
    
        
    for i in range(1, 4):
        if (melody_to_analyze_df.Bir.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.Bir.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note
        
    for i in range(4):    
        if (melody_to_analyze_df.ve.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.ve.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note
   
        if (melody_df.Ki.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.Ki.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note
            
        if (melody_df.ve2.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.ve2.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note        


    
    
    return max_note-min_note+1
```

```python
def check_es(notes):
    
    es_length = "nan"
    if (notes.iloc[0]["Bir"] == "nan") & (notes.iloc[0]["ve"] == "nan") :
        es_length = "Quarter"
    elif notes.iloc[0]["Bir"] == "nan":
        es_length = "Eigth"
    
    return es_length 
 
    
```

```python
def measure_conversion(notes, s_measure):
    
    # x _ _ _
    if (notes.iloc[0]["ve"] == "nan") & (notes.iloc[0]["Ki"] == "nan") & (notes.iloc[0]["ve2"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Half"], 1)
    
    # x _ _ x
    elif (notes.iloc[0]["ve"] == "nan") & (notes.iloc[0]["Ki"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["QuarterDot"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
    
    # x _ x -
    elif (notes.iloc[0]["ve"] == "nan") & (notes.iloc[0]["ve2"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Quarter"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Quarter"], 1)
    
    
    # x _ x x
    elif (notes.iloc[0]["ve"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Quarter"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
    
    # x x _ _
    elif (notes.iloc[0]["Ki"] == "nan") & (notes.iloc[0]["ve2"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["QuarterDot"], 1)
    
    # x x _ x
    elif (notes.iloc[0]["Ki"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["Quarter"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
    
    # x x x _
    elif (notes.iloc[0]["ve2"] == "nan"):           
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Quarter"], 1)
    
    # x x x x
    else:
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
        

    return s_measure
        
```

```python
def write_score(melody):
    
    
    s_temp = stream.Stream()
    first_measure = melody.iloc[[0]]
    second_measure = melody.iloc[[1]]
    third_measure = melody.iloc[[2]]
    
    measure_1_2 = check_es(second_measure)
    measure_2_3 = check_es(third_measure)
    

    # First measure x _ _ _
    
    if (first_measure.iloc[0]["ve"] == "nan") & (first_measure.iloc[0]["Ki"] == "nan") & (first_measure.iloc[0]["ve2"] == "nan"):
        note_add = note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Half"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)      
            
            
    # First measure x _ _ x
            
    elif (first_measure.iloc[0]["ve"] == "nan") & (first_measure.iloc[0]["Ki"] == "nan"):
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["QuarterDot"],1)
        note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)      

    # First measure x _ x _
    
    elif (first_measure.iloc[0]["ve"] == "nan") & (first_measure.iloc[0]["ve2"] == "nan"):
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Quarter"],1)
        note_add = note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Quarter"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)      
        
    # First measure x _ x x   

    elif (first_measure.iloc[0]["ve"] == "nan"):
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Quarter"],1)
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Eigth"],1)
        note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)   
    
    # First measure x x _ _
    
    elif (first_measure.iloc[0]["Ki"] == "nan") & (first_measure.iloc[0]["ve2"] == "nan"):
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Eigth"],1)
        note_add = note_conversion_df.loc[first_measure.iloc[0]["ve"]]["QuarterDot"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)      

    # First measure x x _ x        
            
    elif (first_measure.iloc[0]["Ki"] == "nan"):
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Eigth"],1)
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["ve"]]["Quarter"],1)
        note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)      

    # First measure x x x _        
            
    elif (first_measure.iloc[0]["ve2"] == "nan"):           
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Eigth"],1)
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["ve"]]["Eigth"],1)
        note_add = note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Quarter"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)      

    # First measure x x x x
                
    else:
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Bir"]]["Eigth"],1)
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["ve"]]["Eigth"],1)
        s_temp.repeatAppend(note_conversion_df.loc[first_measure.iloc[0]["Ki"]]["Eigth"],1)
        note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
        if (measure_1_2=="nan"):
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Eigth"):
           
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Eigth"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)
        elif (measure_1_2=="Quarter"):
            
            note_add.tie = tie.Tie("start")
            s_temp.repeatAppend(note_add,1)
            note_add = note_conversion_df.loc[first_measure.iloc[0]["ve2"]]["Quarter"]
            note_add.tie = tie.Tie("stop")
            s_temp.repeatAppend(note_add,1)      

# 2nd Measure        


    # Seconda measure x _ _ _   

    if (measure_1_2=="nan"):
        if (second_measure.iloc[0]["ve"] == "nan") & (second_measure.iloc[0]["Ki"] == "nan") & (second_measure.iloc[0]["ve2"] == "nan"):
            note_add = note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Half"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      
            
            
        # Seconda measure x _ _ x
            
        elif (second_measure.iloc[0]["ve"] == "nan") & (second_measure.iloc[0]["Ki"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["QuarterDot"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

        
        # Seconda measure x _ x _
        
        elif (second_measure.iloc[0]["ve"] == "nan") & (second_measure.iloc[0]["ve2"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Quarter"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
               
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      
        
        # Seconda measure x _ x x
        
        elif (second_measure.iloc[0]["ve"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Quarter"],1)
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
               
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)   
        
        # Seconda measure x x _ _
        
        elif (second_measure.iloc[0]["Ki"] == "nan") & (second_measure.iloc[0]["ve2"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Eigth"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve"]]["QuarterDot"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
               
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

         # Second measure x x _ x
        
        elif (second_measure.iloc[0]["Ki"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Quarter"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

         # Second measure x x x _  
            
        elif (second_measure.iloc[0]["ve2"] == "nan"):           
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Eigth"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

        # Second measure x x x x 
        
        else:
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Bir"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

    # ilk Nota Eight Es
    
    elif (measure_1_2=="Eigth"):
        
        #2nd measure _ x _ _
        
        if (second_measure.iloc[0]["Ki"] == "nan") & (second_measure.iloc[0]["ve2"] == "nan"):
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve"]]["QuarterDot"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

        #2nd measure _ x _ x 
        
        elif (second_measure.iloc[0]["Ki"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Quarter"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

        #2nd measure _ x x _    
            
        elif (second_measure.iloc[0]["ve2"] == "nan"):           
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Eigth"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

        #2nd measure _ x x x        
            
        else:
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["ve"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"],1)
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      
        
    # ilk Nota Quarter Es
    
    elif (measure_1_2=="Quarter"):
        
        #2nd measure _ _ x _ 
        
        if (second_measure.iloc[0]["ve2"] == "nan"):           
            note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)      

        #2nd measure _ _ x x        
            
        else:
           
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["Ki"]]["Eigth"],1)
            #note_add.tie = tie.Tie("stop")
            note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
            #note_add.tie = tie.Tie("stop")
            if (measure_2_3=="nan"):
                s_temp.repeatAppend(note_add,1)
            elif (measure_2_3=="Eigth"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
                
            elif (measure_2_3=="Quarter"):
                
                note_add.tie = tie.Tie("start")
                s_temp.repeatAppend(note_add,1)
                note_add = note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Quarter"]
                note_add.tie = tie.Tie("stop")
                s_temp.repeatAppend(note_add,1)
                
        
# 3rd measure

  
    #Third Measure x _ _ _  
    
    if (measure_2_3=="nan"):
        if (third_measure.iloc[0]["ve"] == "nan") & (third_measure.iloc[0]["Ki"] == "nan") & (third_measure.iloc[0]["ve2"] == "nan"):
            note_add = note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["Half"]
            s_temp.repeatAppend(note_add,1)
              
         #Third Measure x _ _ x   
        
        elif (third_measure.iloc[0]["ve"] == "nan") & (third_measure.iloc[0]["Ki"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["QuarterDot"],1)
            note_add = note_conversion_df.loc[third_measure.iloc[0]["ve2"]]["Eigth"]
            s_temp.repeatAppend(note_add,1)
             
        #Third Measure x _ x _ 
            
        elif (third_measure.iloc[0]["ve"] == "nan") & (third_measure.iloc[0]["ve2"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["Quarter"],1)
            note_add = note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Quarter"]
            s_temp.repeatAppend(note_add,1)
              
        #Third Measure x _ x x
        
        elif (third_measure.iloc[0]["ve"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["Quarter"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Eigth"],1)
            note_add = note_conversion_df.loc[third_measure.iloc[0]["ve2"]]["Eigth"]
            s_temp.repeatAppend(note_add,1)
         
        #Third Measure x x _ _
            
        elif (third_measure.iloc[0]["Ki"] == "nan") & (third_measure.iloc[0]["ve2"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["Eigth"],1)
            note_add = note_conversion_df.loc[third_measure.iloc[0]["ve"]]["QuarterDot"]
            s_temp.repeatAppend(note_add,1)
            
        #Third Measure x x _ x
        
        elif (third_measure.iloc[0]["Ki"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve"]]["Quarter"],1)
            note_add = note_conversion_df.loc[third_measure.iloc[0]["ve2"]]["Eigth"]
            s_temp.repeatAppend(note_add,1)
            
        #Third Measure x x x -   
            
        elif (third_measure.iloc[0]["ve2"] == "nan"):           
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve"]]["Eigth"],1)
            note_add = note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Quarter"]
            s_temp.repeatAppend(note_add,1)
        
        #Third Measure x x x x 
        else:
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Bir"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve2"]]["Eigth"],1)
          
            
    # ilk Nota Eight Es
    
    elif (measure_2_3=="Eigth"):
        
        #Third Measure _ x _ _  
        
        if (third_measure.iloc[0]["Ki"] == "nan") & (third_measure.iloc[0]["ve2"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve"]]["QuarterDot"],1)
          
         
        #Third Measure _ x _ x  
            
        elif (third_measure.iloc[0]["Ki"] == "nan"):
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve"]]["Quarter"],1)
            s_temp.repeatAppend(note_conversion_df.loc[second_measure.iloc[0]["ve2"]]["Eigth"],1)
             

        #Third Measure _ x x _      
            
        elif (third_measure.iloc[0]["ve2"] == "nan"):           
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Quarter"],1)
                         
        #Third Measure _ x x x  
            
        else:
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve2"]]["Eigth"],1)
                   
        
    # ilk Nota Quarter Es
    
    elif (measure_2_3=="Quarter"):
        
       #Third Measure _ _ x _  
    
        if (third_measure.iloc[0]["ve2"] == "nan"):           
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Quarter"],1)
         
        #Third Measure _ _ x x     
        else:
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["Ki"]]["Eigth"],1)
            s_temp.repeatAppend(note_conversion_df.loc[third_measure.iloc[0]["ve2"]]["Eigth"],1)
   
       
        
        
        
    s_temp = measure_conversion(melody.iloc[[3]], s_temp)  
    s_temp.repeatAppend(w_C, 1)
    s_temp.insert(0, meter.TimeSignature('4/4'))    
        
    return s_temp
    
```
