```python
def total_range(melody_to_analyze_df):

    min_note = note_conversion_df.Order[melody_to_analyze_df.Bir.iloc[0]]
    max_note = note_conversion_df.Order[melody_to_analyze_df.Bir.iloc[0]]
    
        
    for i in range(1, 4):
        if (melody_to_analyze_df.Bir.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.Bir.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note
        
    for i in range(4):    
        if (melody_to_analyze_df.ve.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.ve.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note
   
        if (melody_df.Ki.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.Ki.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note
            
        if (melody_df.ve2.iloc[i]!="nan"):
            temp_note = note_conversion_df.Order[melody_to_analyze_df.ve2.iloc[i]]
            if temp_note>max_note:
                max_note = temp_note
            if temp_note<min_note:
                min_note = temp_note        


    
    
    return max_note-min_note+1
```

```python
def check_es(notes):
    
    es_length = "nan"
    if (notes.iloc[0]["Bir"] == "nan") & (notes.iloc[0]["ve"] == "nan") :
        es_length = "Quarter"
    elif notes.iloc[0]["Bir"] == "nan":
        es_length = "Eigth"
    
    return es_length 
 
    
```

```python
def measure_conversion(notes, s_measure):
    
    # x _ _ _
    if (notes.iloc[0]["ve"] == "nan") & (notes.iloc[0]["Ki"] == "nan") & (notes.iloc[0]["ve2"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Half"], 1)
    
    # x _ _ x
    elif (notes.iloc[0]["ve"] == "nan") & (notes.iloc[0]["Ki"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["QuarterDot"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
    
    # x _ x -
    elif (notes.iloc[0]["ve"] == "nan") & (notes.iloc[0]["ve2"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Quarter"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Quarter"], 1)
    
    
    # x _ x x
    elif (notes.iloc[0]["ve"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Quarter"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
    
    # x x _ _
    elif (notes.iloc[0]["Ki"] == "nan") & (notes.iloc[0]["ve2"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["QuarterDot"], 1)
    
    # x x _ x
    elif (notes.iloc[0]["Ki"] == "nan"):
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["Quarter"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
    
    # x x x _
    elif (notes.iloc[0]["ve2"] == "nan"):           
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Quarter"], 1)
    
    # x x x x
    else:
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Bir"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["Ki"]]["Eigth"], 1)
        s_measure.repeatAppend(note_conversion_df.loc[notes.iloc[0]["ve2"]]["Eigth"], 1)
        

    return s_measure
        
```
